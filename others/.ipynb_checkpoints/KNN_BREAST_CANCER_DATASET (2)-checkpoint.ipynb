```python
from sklearn import datasets,metrics
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
cancerData = datasets.load_breast_cancer()
```

```python
# Splitting data
X = pd.DataFrame(data = cancerData.data,
                 columns = cancerData.feature_names)
X.head()
```

```python
X.shape
```

```python
y = pd.DataFrame(cancerData.target, columns =['diagnosis'])
y.shape
```

# Exploring the data

```python
y.value_counts()
```

```python
data = pd.concat([X,y], axis=1)
```

```python
sns.set_style('whitegrid')
# color palettes
pal = ['#0e2433', '#ff007f']
sns.countplot(data['diagnosis'],palette = pal,alpha = 0.8)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
m = ax.hist(data[data["diagnosis"] == 0]['mean radius'], bins=20,range=(0, 30), 
            label = "Malignant", alpha=0.7, color='#232121')
b = ax.hist(data[data["diagnosis"] == 1]['mean radius'], bins=20,range=(0, 30),
            label = "Benign", alpha=0.7, color='#df2378')
plt.xlabel("Radius")
plt.ylabel("Count")
plt.title("Mean Radius")
plt.legend()
plt.show()
```

```python
#Splitting training and testing data
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.2, stratify = y)
```

```python
X_train.shape
```

```python
X_test[y_test == 0]
```

```python
for i in X_test.columns:
    print( "<th scope='col'>{}</th>".format(i.capitalize()))
```

```python
import numpy as np
row1 =  [np.round(x,2) for x in X_test[y_test == 0].values[2]]
for i in row1:
    print( "<td scope='col'>{}</td>".format(i))
```

```python
import numpy as np
for i in range(3):
    print( "<td scope='row'>{}</td>".format(i+1))
    row1 =  [np.round(x,2) for x in X_test[y_test == 1].values[i]]
    for i in row1:
        print( "<td scope='col'>{}</td>".format(i))
```

```python
# Scaling the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
# X_train = X_train.reshape(512, 8, 1)
# X_test = X_test.reshape(57,8,1)
```

```python
X_train[0]
```

```python
y
```

```python
# Creating the model
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train, y_train)

pred = model.predict(X_test)

```

```python
# Accuracy Measures
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report

print(accuracy_score(pred,y_test))
```

```python
# Elbow method
import numpy as np
error_rate = []

for i in range(1,40):

  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  pred_i = knn.predict(X_test)
  error_rate.append(np.mean(pred_i!= y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
 markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
range(1,40)[error_rate.index(min(error_rate))]
```

```python
model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
# Accuracy Measures
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report

print(accuracy_score(pred,y_test))
```

```python
conf = confusion_matrix(pred, y_test)

sns.heatmap(conf, annot = True, fmt = '.2f')
```

```python
print(classification_report(pred, y_test))
```

```python
from sklearn.metrics import precision_recall_fscore_support

precision = precision_recall_fscore_support(y_test,pred, average = 'macro') 
print("Precision :",precision)
```

```python
from collections import Counter
from sklearn.metrics import roc_curve,auc,roc_auc_score 
fpr, tpr, thresh =roc_curve(y_test, pred, pos_label = None)
print('ROC : ', auc(fpr,tpr))
```

```python
roc_log = roc_auc_score(y_test, pred)
auc = auc(fpr,tpr)
plt.plot([0, 1], [0, 1], 'r--')
plt.plot(fpr, tpr, label='AUC = {:.3f}'.format(auc))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
plt.close()
```
